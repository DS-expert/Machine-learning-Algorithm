```python
# Encode all the features
X  = df.drop("class", axis=1)
y = df["class"]
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

one_hot = OneHotEncoder(sparse_output=False, drop="first", handle_unknown="ignore")

X_encoded = one_hot.fit_transform(X)
```

```python
label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
class LogisticRegression():

    def __init__(self, learning_rate, epochs):

        self.coef_ = None
        self.lr = learning_rate
        self.epochs = epochs
    
    def sigmoid(self, z):

        return 1 / (1 + np.exp(-z))
    
    def fit(self, X_train, y_train):

        X_train = np.insert(X_train, 0, 1, axis=1)
        self.coef_ = np.random.randn(X_train.shape[1]) * 0.01

        for i in range(self.epochs):

            y_hat = self.sigmoid(np.dot(X_train, self.coef_))

            error = y_hat - y_train

            coef_slope = np.dot(X_train.T, error) / X_train.shape[0]

            self.coef_ = self.coef_ - (self.lr * coef_slope)
        
        print(f"Coef_: {self.coef_}")
    
    def predict(self, X_test):

        X_test = np.insert(X_test, 0, 1, axis=1)
        y_pred = self.sigmoid(np.dot(X_test, self.coef_))

        return (y_pred >= 0.5).astype(int)
```

```python
logistc_ = LogisticRegression(learning_rate=0.001, epochs=1000)
```

```python
logistc_.fit(X_train, y_train)
```

```python
y_pred_logistic = logistc_.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy_logistic = accuracy_score(y_pred_logistic, y_test)
classification_report_logistic = classification_report(y_pred_logistic, y_test)
confustion_matrix_logistic = confusion_matrix(y_pred_logistic, y_test)

print(f"Accuracy Logistic : {accuracy_logistic * 100:.2f}%")
print(f"Classification Report Logistic : \n{classification_report_logistic}")
print(f"Confusion Matrix Logistic : \n{confustion_matrix_logistic}")
```

```python
# Decision Tree Classifier 

class DecisionTree_():

    def __init__(self, max_depth=None):

        self.max_depth = max_depth
        self.tree = None
    
    def gini(self, y_train):

        classes, counts = np.unique(y_train, return_counts=True)

        p = counts / counts.sum()

        return 1 - np.sum(p ** 2)

    def best_split(self, X_train, y_train):

        best_gini = float("inf")
        best_feature = None
        best_value = None
        n_samples, n_features = X_train.shape

        for feature in range(n_features):

            values = np.unique(X_train[:, feature])

            for value in values:

                left_mask = X_train[:, feature] <= value
                right_mask = X_train[:, feature] > value

                y_left = y_train[left_mask]
                y_right = y_train[right_mask]

                gini_left = self.gini(y_left)
                gini_right = self.gini(y_right)

                weighted_gini = (len(y_left) * gini_left + len(y_right) * gini_right) / n_samples

                if weighted_gini < best_gini:

                    best_gini = weighted_gini
                    best_feature = feature
                    best_value = value
        
        return best_feature, best_value

    def build_tree(self, X_train, y_train, depth=0):

        class_counts = np.bincount(y_train)
        majorities_classes = np.argmax(class_counts)

        if len(set(y_train)) == 1 or len(y_train) == 0 or (self.max_depth is not None and depth >= self.max_depth):

            return {"leaf": True, "class" : majorities_classes}

        best_features, best_values = self.best_split(X_train, y_train)

        if best_features is None:

            return {"leaf": True, "class" : majorities_classes}

        left_mask = X_train[:, best_features] <= best_values
        right_mask = X_train[: ,best_features] > best_values

        left_tree = self.build_tree(X_train[left_mask], y_train[left_mask], depth + 1)
        right_tree = self.build_tree(X_train[right_mask], y_train[right_mask], depth + 1)

        return {
            "leaf" : False,
            "feature" : best_features,
            "value" : best_values,
            "left": left_tree,
            "right" : right_tree
        }

    def fit(self, X_train, y_train):
        
        self.tree = self.build_tree(X_train, y_train)
    
    def predict_one(self, X_test):

        node = self.tree

        while isinstance(node, dict) and not node["leaf"]:

            feature_ = node["feature"]
            value_ = node["value"]

            if X_test[feature_] <= value_:

                node = node["left"]
            
            else:

                node = node["right"]

        return node["class"]

    def predict(self, X_test):

        return np.array([self.predict_one(x) for x in X_test])
```

```python
decision_ = DecisionTree_(max_depth=5)
```

```python
decision_.fit(X_train, y_train)
```

```python
y_pred_decision = decision_.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_pred_decision, y_test, cmap="Blues")
plt.title(f"Confusion Matrix Decision Tree")
plt.show()
```
