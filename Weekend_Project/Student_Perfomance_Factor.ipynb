```python
# Remove Outliers

def remove_outliers(df, feature_name ):

    """
    Calculate the lower bound and upper bound using IQR and using ArbitaryOutlierCapper
    remove the outliers from that specific feature
    *param:dataset (df), feature_name
    *return: lower_bound, upper_bound
    """
    q1 = df[feature_name].quantile(0.25)
    q3 = df[feature_name].quantile(0.75)

    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    from feature_engine.outliers import ArbitraryOutlierCapper

    arb = ArbitraryOutlierCapper(min_capping_dict={feature_name:lower_bound}, max_capping_dict={feature_name:upper_bound})

    df = arb.fit_transform(df)

    return df
```

```python
df = remove_outliers(df, "Hours_Studied")
```

```python
df = remove_outliers(df, "Tutoring_Sessions")
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(df)
plt.show()
```

```python
# Split dataset into Train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("Exam_Score", axis=1), df["Exam_Score"], test_size=0.2, random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
y_train
```

```python
# Encoding Categorical Features
from sklearn.preprocessing import OneHotEncoder

encoder_ = OneHotEncoder(sparse_output=False, drop="first", handle_unknown="ignore")

X_train = encoder_.fit_transform(X_train)
X_test = encoder_.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lr_ = LinearRegression()

lr_.fit(X_train, y_train)
```

```python
y_pred_ = lr_.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2_score_lr = r2_score(y_test, y_pred_)
mse_lr = mean_squared_error(y_test, y_pred_)
rmse_lr = np.sqrt(mse_lr)

print(f"R2 Score LR: {r2_score_lr:.2f}%")
print(f"MSE LR: {mse_lr:.2f}")
print(f"RMSE LR: {rmse_lr:.2f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_transformer = PolynomialFeatures()

X_train_poly = poly_transformer.fit_transform(X_train)
X_test_poly = poly_transformer.transform(X_test)
```

```python
y_pred_poly = lr_poly.predict(X_test)
```

```python
r2_score_poly = r2_score(y_test, y_pred_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)
rmse_poly = np.sqrt(mse_poly)

print(f"R2 Score Poly: {r2_score_poly:.2f}%")
print(f"MSE Poly: {mse_poly:.2f}")
print(f"RMSE Poly: {rmse_poly:.2f}")
```
