```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
# Load the data
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                          n_clusters_per_class=1, random_state=50, class_sep=10)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=50)
plt.show()
```

```python
def sigmoid(z):

    return 1 / (1 + np.exp(-z))
```

```python
def perceptron(X,y):

    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.01

    for i in range(1000):

        j = np.random.randint(X.shape[0])
        y_hat = sigmoid(np.dot(X[j], weights))
        weights = weights - (lr * (y[j] - y_hat) * X[j])
    
    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
intercept_, coef_
```

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_ / coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color="r", linewidth=2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter")
plt.show()
```
