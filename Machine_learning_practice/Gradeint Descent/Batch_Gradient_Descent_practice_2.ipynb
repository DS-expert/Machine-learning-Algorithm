```python
class Batch_Gradient_Descent():

    def __init__(self, learning_rate, epochs):
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
    
    def fit(self, X_train, y_train):
        X_train = np.insert(X_train, 0,1, axis=1)
        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_ = 1

        for i in range(self.epochs):
            y_hat = self.intercept_ + np.dot(X_train, self.coef_)
            intercept_slope = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_ + (self.lr * intercept_slope)

            coef_slope = -2 * np.mean(np.dot((y_train - y_hat), X_train))
            self.coef_ = self.coef_ + (self.lr * coef_slope)
        
        print(f"Intercept value: {self.intercept_}, Coef Values: {self.coef_}")
    
    def predict(self, X_test):
        X_test = np.insert(X_test, 0, 1, axis=1)
        y_pred = np.dot(X_test, self.coef_) + self.intercept_
        return y_pred

    
```

```python
bgd = Batch_Gradient_Descent(learning_rate=0.0001, epochs=20)
```

```python
bgd.fit(X_train, y_train)
```

```python
y_pred = bgd.predict(X_test)
```

```python
def r2_score_custom(y_true, y_pred):

    mean_value = np.mean(y_true)

    SSE = np.sum((y_true - y_pred) ** 2)
    TSS = np.sum((y_true - mean_value) ** 2)

    r2_score = 1 - (SSE / TSS)
    return r2_score
```

```python
r2_score_c = r2_score_custom(y_test, y_pred)
r2_score_c
```
