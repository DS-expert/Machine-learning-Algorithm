```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, hypercube=False, 
                           random_state=42)
```

```python
X.shape, y.shape
```

```python
y
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x=X[:, 0], y=X[:, 1], c=y, cmap="winter", s=100)
```

```python
def step(z):

    return 1 if z > 0 else 0
```

```python
y
```

```python
def perceptron(X, y):

    X = np.insert(X, 0, 1, axis=1)
    weight = np.ones(X.shape[1])
    lr = 0.01

    for i in range(1000):

        j = np.random.randint(X.shape[0])
        y_hat = step(np.dot(X[j], weight))
        weight = weight + lr * (y[j] * y_hat) * X[j]
    
    return weight[0], weight[1:]
```

```python
y
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
y
```

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_ / coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color="r", linewidth=2)
plt.scatter(x=X[:, 1], y=X[:, 0], c=y, cmap="winter", s=100)
plt.ylim(-3, 2)
plt.show()
```
