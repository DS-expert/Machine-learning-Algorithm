```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 10)
```

```python
x
```

```python
plt.plot(x)
```

```python
def sigmoid(z):

    return 1 / (1 + np.exp(-z))
```

```python
sigmoid(x)
```

```python
plt.plot(sigmoid(x))
```

```python
def derivative_sigmoid(z):

    return np.dot(sigmoid(z), (1 - sigmoid(z)))
```

```python
derivative_sigmoid(x)
```

```python
plt.plot(sigmoid(x))
```

```python
derivative_sigmoid(sigmoid(x))
```
